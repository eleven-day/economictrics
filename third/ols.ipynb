```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'economics3.xlsx',sheet_name=3,header=2)
df_list = df.values.tolist()

x = []
y = []

for i in df_list:
    x.append(i[0])
for i in df_list:
    y.append(i[1])
    
x = np.array(x)
y = np.array(y)

def ols(x,y):
    a = (len(x)*sum(x*y) - sum(x)*sum(y))/(len(x)*sum(x**2) - sum(x)*sum(x))
    b = (sum(y) - a*sum(x))/len(x)
    r2 = ((len(x)*sum(x*y) - sum(x)*sum(y))**2)/((len(x)*sum(x**2) - sum(x)*sum(x))*(len(y)*sum(y**2) - sum(y)*sum(y)))
    return a,b,r2
print(ols(x,y))
```
