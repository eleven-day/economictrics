```python
#几何平均
import pandas as pd

def geometric_mean(data):  # 计算几何平均数
    total=1
    for i in data:
        total*=i #等同于total=total*i
    return pow(total,1/len(data))

a = [5.4,4.9,1.3,4.6,5.0]
 
a = list((pd.Series(a) + 100)/100)

print(geometric_mean(a))
```

```python
#移动平均
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

x = range(1980,1996)
y = [1474,1406,784,875,1200,1752,2568,2303,2687,2837,3025,4297,5512,11888,8191,10073]

def movingAverage(signal, window):

    sum = 0
    mAver = []
    k = int((window-1)/2)

    for i in np.arange(k, len(signal)-k):
        for ii in np.arange(i-k, i+k+1):
            sum = sum + signal[ii]
        mAver.append(sum / window)
        sum = 0
    return mAver

xm = range(1981,1995)
ym = movingAverage(y, 3)

plt.plot(x,y)
plt.plot(xm,ym)
```
