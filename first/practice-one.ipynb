```python
import numpy as np

Y0 = np.array([12361,271,570,2023,4287,5630,17948,114333,11240,10812])
Yg = np.array([12557,299,619,2122,4485,5946,19375,121121,13025,12043])

def geometric_mean(data):  # 计算几何平均数
    total=1
    for i in data:
        total*=i #等同于total=total*i
    return pow(total,1/len(data))

Y = pow((Yg/Y0),1/5)

geometric_mean(Y)
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

O = np.array([26.20,18.00,17.00,15.50,20.30,27.80,17.90,18.30,13.25,16.55,18.35])
G = np.array([327.00,390.90,486.50,410.15,398.60,391.00,353.40,332.90,390.65,382.50,386.70])

def movingAverage(signal, window):

    sum = 0
    mAver = []
    k = int((window-1)/2)

    for i in np.arange(k, len(signal)-k):
        for ii in np.arange(i-k, i+k+1):
            sum = sum + signal[ii]
        mAver.append(sum / window)
        sum = 0
    return mAver

x1 = np.arange(1986,1995)
y1 = movingAverage(O,3)
plt.plot(x1,y1)
print(y1)
```

```python
x2 = np.arange(1986,1995)
y2 = movingAverage(G,3)
plt.plot(x2,y2)
print(y2)
```

```python
N = np.array([3586,7662,4358,4034,3743,2755,7912])
C = np.array([8.79,19.11,10.89,6.31,9.76,7.17,14.36])

def R(x,y):
    a = x.sum()
    b = y.sum()
    c = sum(x*y)
    d = sum(x*x)
    e = sum(y*y)
    f = len(x)

    R = (f*c - a*b)/(((f*d - a*a)*(f*e - b*b))**(1/2))
    return R
R(N,C)
```
